```python
x = [4,7,9,10]
y = [25,90, 20,40]
plt.plot(x, y, label="legend")

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('halsdfa')

plt.legend()

plt.yticks([0,7,8], ['43', '3', '%'])

plt.show()
```
